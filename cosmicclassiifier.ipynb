```python
missing_count = df["Magnetic Field Strength"].isnull().sum()
random_samples = np.random.choice(df["Magnetic Field Strength"].dropna(), size=missing_count, replace=True)
df.loc[df["Magnetic Field Strength"].isnull(), "Magnetic Field Strength"] = random_samples
```

```python
missing_count = df['Radiation Levels'].isnull().sum()
random_samples = np.random.choice(df['Radiation Levels'].dropna(), size=missing_count, replace=True)
df.loc[df["Radiation Levels"].isnull(), "Radiation Levels"] = random_samples
```

so, mean imputation is causing bias.

instead we use random sampling.

```python
import numpy as np

# Select only numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Apply random sampling imputation to all numeric columns
for col in numeric_cols:
    missing_indices = df[col].isnull()
    df.loc[missing_indices, col] = np.random.choice(df[col].dropna(), size=missing_indices.sum())
```

```python
col = "Orbital Period"  # Specify the column

plt.figure(figsize=(12, 5))

# Plot before imputation
plt.subplot(1, 2, 1)
df_missing[col].hist(bins=30, edgecolor='black')
plt.title(f'Before Imputation: {col}')

# Plot after imputation
plt.subplot(1, 2, 2)
df[col].hist(bins=30, edgecolor='black')
plt.title(f'After Imputation: {col}')
```

```python
X = df.drop(columns=["Prediction"])
y = df["Prediction"]
```

```python
numeric_cols = X.select_dtypes(include=['number']).columns
categorical_cols = X.select_dtypes(exclude=['number']).columns

z_scores = zscore(X[numeric_cols])

# Create a mask to filter out outliers
mask = (z_scores > -3.5) & (z_scores < 3.5)

outliers = ((z_scores > 3.5) | (z_scores < -3.5)).sum()
print(outliers)

X_cleaned = X[mask.all(axis=1)].copy()
y_cleaned = y.loc[X_cleaned.index]

print(f"\nOriginal shape: {X.shape}")
print(f"New shape after outlier removal: {X_cleaned.shape}")
```

```python
from scipy.stats import skew
skew_values = X[numeric_cols].apply(skew)
print(skew_values)
```

```python
X_cleaned['Magnetic Field Strength'] = X_cleaned['Magnetic Field Strength'].str.replace('Category_', '', regex=True).astype(float)
X_cleaned['Radiation Levels'] = X_cleaned['Radiation Levels'].str.replace('Category_', '', regex=True).astype(float)
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y_cleaned, test_size=0.1, random_state=42)

# XGBClassifier
xgb_clf = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    n_estimators=500,
    learning_rate=0.2,
    max_depth=15,
    subsample=0.8,
    colsample_bytree=0.8
)

xgb_clf.fit(X_train, y_train)

y_pred = xgb_clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"{accuracy:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest Classifier
rf_clf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)

rf_clf.fit(X_train, y_train)

y_pred = rf_clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"{accuracy:.4f}")
```

```python
from sklearn.naive_bayes import GaussianNB

nb_classifier = GaussianNB()

nb_classifier.fit(X_train, y_train)
y_pred = nb_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```
